```python
path="/Users/alexander/Documents/Inteligentes/proyecto final/competencia/imgset"
img_path="/Users/alexander/Documents/Inteligentes/proyecto final/competencia/test"
inputt="/Users/alexander/Documents/Inteligentes/proyecto final/test.csv"

create_directory(path)

```

```python
relu=load_net("/Users/alexander/Documents/Inteligentes/proyecto final/competencia/model_rel_m.h5")
cnn=load_net("/Users/alexander/Documents/Inteligentes/proyecto final/competencia/cnn_mod.h5")
```

```python
X=separate_images(inputt,path,img_path)
```

```python


data=[]
for img in X:
    name=img
    img = image.load_img(os.path.join(path,img), target_size=(256, 256))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    preds = cnn.predict(x)
    Y=relu.predict_classes(preds)
    tup=(name,np.array2string(Y).lstrip('[').rstrip(']'))
    data.append(tup)
    

```
